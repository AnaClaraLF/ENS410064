```python
import pandas as pd
import calendar
import datetime
import matplotlib.pyplot as plt

#Criando um DataFrame pandas com a tabela csv que baixamos.
#Info:
#delimiter - Delimitador da tabela csv
#enconding - A codificação da tabela
#index_col - Primeira Coluna não é index da tabela (False)
#skiprows - Pular as 13 linhas iniciais para formar a tabela (Evitando #o cabeçalho)
df=pd.read_csv(r'D:\PPGEA\AulaPy\DataENS\rawDataHidroweb\65028000\VAZOES.txt',
               delimiter=';', encoding='ISO-8859-1',index_col=False,skiprows=16)

#Alterar separador decimal
df=df.replace(',', '.', regex = True)

Lista_de_dados=[ 'Data', 'Vazao01', 'Vazao02', 'Vazao03',
       'Vazao04', 'Vazao05', 'Vazao06', 'Vazao07', 'Vazao08', 'Vazao09',
       'Vazao10', 'Vazao11', 'Vazao12', 'Vazao13', 'Vazao14', 'Vazao15',
       'Vazao16', 'Vazao17', 'Vazao18', 'Vazao19', 'Vazao20', 'Vazao21',
       'Vazao22', 'Vazao23', 'Vazao24', 'Vazao25', 'Vazao26', 'Vazao27',
       'Vazao28', 'Vazao29', 'Vazao30', 'Vazao31']

#Apenas os dados da Lista
df=df[Lista_de_dados]

#Redefinir a coluna 'Data' com formato "texto" para "data"
df['Data']=pd.to_datetime(df['Data'], format='%d/%m/%Y')

#Listas vazias
Lista_Vazao=[]
Lista_Data=[]

#Iteração a partir do comprimento total da tabela
#Para cada coluna Data, obtem-se o ano e o mes
for n in range(0,len(df)):
    
    year_info=df['Data'][n].year
    month_info=df['Data'][n].month   
    
    #Nessa iteração utilizamos a biblioteca calendar
    #para extrair qual é a data do último dia do mês
    #Assim podemos fazer a iteração sem errar a localizacao
    # de cada valor diário de vazão
    for day in range(0,calendar.monthrange(year_info, month_info)[1]):
            VAZAO=df.iloc[[n]][df.columns[day+1]]
        	
            #Bliblioteca Datetime para criar
            #A data da vazao year/month/day
            DATA=datetime.datetime(year_info, month_info, day+1)
            
            #Adicionar os resultados na lista
            Lista_Vazao.append(float(VAZAO.values[0]))
            Lista_Data.append(DATA)

#Por fim, basta criar a nossa tabela final 

Tabela_Final=pd.DataFrame(Lista_Data, columns=['Data'])
Tabela_Final['Vazao']=Lista_Vazao


fig, ax = plt.subplots(1,1,figsize=(12, 5), dpi= 300) 

#Parametros do gráfico 
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['axes.facecolor'] = 'white'

##Gridlines
#plt.grid(b=False, which='major', color='#666666', linestyle='-')
#plt.minorticks_on()
#plt.grid(b=True, which='minor', color='#999999', linestyle='-', 
#    		alpha=0.3)

#Legenda eixo x
plt.xlabel('Data')

#Título
plt.title('Estação: 65028000')

#Plot
plt.plot(Tabela_Final['Data'],Tabela_Final['Vazao'],label='Vazão m³/s',linewidth=0.3,color='blue')

plt.legend()
```
